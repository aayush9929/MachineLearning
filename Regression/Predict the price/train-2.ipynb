```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("dataset/train.csv")
```

```python
data
```

```python
data.info()
```

```python
data['instock_date'] = pd.to_datetime(data['instock_date'])
```

```python
data.info()
```

```python
data['instock_month'] = data['instock_date'].dt.month
data['instock_year'] = data['instock_date'].dt.year
```

```python
data.info()
```

```python
data.drop(columns=['instock_date'],inplace=True)
```

```python
data.info()
```

```python
data.drop(columns=['Customer_name'],inplace=True)
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lo_cu = LabelEncoder()
pr_ca = LabelEncoder()
data['loyalty_customer_encoding'] = lo_cu.fit_transform(data['Loyalty_customer'])
data['product_category_encoding'] = pr_ca.fit_transform(data['Product_Category'])
```

```python
data.info()
```

```python
data.drop(columns=['Loyalty_customer','Product_Category'],inplace=True)
```

```python
data.info()
```

```python
for i in data.columns:
    print(i)
    if i!='Product_id':
        mean_column = data[i].mean()
        data[i].fillna(mean_column,inplace=True)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X=data.drop(columns=['Product_id','Selling_Price']).values
y=data['Selling_Price'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
reg = LinearRegression().fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,reg.predict(X_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
lr=RandomForestRegressor(max_depth=59)#30,99.5880
lr.fit(X_train,y_train)#19,99.577
r2_score(y_test,lr.predict(X_test))#99.556
```

```python
lr.score(X, y)
```
