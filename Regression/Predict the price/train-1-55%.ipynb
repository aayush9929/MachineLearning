```python
# made it for max price, min price,selling price
for market_category in data.Market_Category.unique():
    replace_value = data.loc[(data['Market_Category']==market_category)].median()['Selling_Price']
    print(replace_value)
    data.loc[(data['Market_Category']==market_category),('Selling_Price')]=data.loc[(data['Market_Category']==market_category),('Selling_Price')].fillna(replace_value)
    print(data.loc[(data['Selling_Price'].isnull()) &  (data['Market_Category']==market_category)])
```

```python
#creating train data
X=data.drop(columns=['Product_id','Selling_Price']).values
y=data['Selling_Price'].values
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=82)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression().fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,reg.predict(X_test))#99.556
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
lr=RandomForestRegressor(max_depth=59)#30,99.5880
```

```python
lr.fit(X_train,y_train)#19,99.577
```

```python
r2_score(y_test,lr.predict(X_test))#99.556
```
